```python
import torch

# Specify the file path
file_path = '/content/drive/MyDrive/TGIF_Features/gif_0.pt'

# Load the object from the file (which seems to be a dictionary)
data = torch.load(file_path)

# Print the dictionary keys to see what it contains
print(f'Keys in the dictionary: {data.keys()}')
```

```python
# Inspect the tensors under each key
act_tensor = data['act']
emo_tensor = data['emo']
obj_tensor = data['obj']

# Print the shape, data type, and device for each tensor
print(f"Action Tensor Shape: {act_tensor.shape}, Data Type: {act_tensor.dtype}, Device: {act_tensor.device}")
print(f"Emotion Tensor Shape: {emo_tensor.shape}, Data Type: {emo_tensor.dtype}, Device: {emo_tensor.device}")
print(f"Object Tensor Shape: {obj_tensor.shape}, Data Type: {obj_tensor.dtype}, Device: {obj_tensor.device}")
```

```python
# Print the first few values of each tensor
print("Action Tensor (first few values):", act_tensor[0, :5])
print("Emotion Tensor (first few values):", emo_tensor[0, :5])
print("Object Tensor (first few values):", obj_tensor[0, :5])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Stack the 'emo' tensor into a 2D array (assuming you have more data for different emotions)
# If you have multiple tensors, you can concatenate them; here, we'll assume you have a set of 'emo' embeddings

# For now, we'll use the same 'emo' tensor repeatedly to simulate multiple embeddings for visualization
# (You would replace this with actual data from your dataset later)
emo_data = emo_tensor.repeat(100, 1)  # Simulating 100 samples with the same tensor

# Apply PCA to reduce from 768 dimensions to 2
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(emo_data)

# Plot the reduced data
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c='blue', label='Emotion Embeddings')
plt.title('PCA of Emotion Embeddings')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
# Check if the values vary across samples
emo_data = emo_tensor.repeat(1000, 1)  # Simulate 1000 samples with the same tensor

# Calculate the variance for each feature (column) across the samples
variance = emo_data.var(dim=0)

# Print the variance for each feature
print("Variance for each feature in the 'emo' tensor:", variance)
```

```python
# Check variance for 'obj' tensor across 1000 samples
obj_data = obj_tensor.repeat(1000, 1)
obj_variance = obj_data.var(dim=0)
print("Variance for each feature in the 'obj' tensor:", obj_variance)
```

```python
# Check variance for 'act' tensor across 1000 samples
act_data = act_tensor.repeat(1000, 1)
act_variance = act_data.var(dim=0)
print("Variance for each feature in the 'act' tensor:", act_variance)
```
